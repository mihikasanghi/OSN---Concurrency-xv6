```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'your_file.csv' with the actual path to your CSV file
file_path = 'val.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df.head()
```

```python
grouped_df = df.groupby('pid')

# Initialize empty lists to store arrays
pid_arrays = []

# Plot each 'pid' on a separate graph
for pid, group in grouped_df:
    time_array = group['time'].tolist()
    priority_array = group['priority'].tolist()

    # Plot on a separate graph for each 'pid'
    plt.figure()
    plt.plot(time_array, priority_array, label=f'pid {pid}')
    plt.xlabel('Time')
    plt.ylabel('Priority')
    plt.title(f'Graph for pid {pid}')
    plt.legend()
    # plt.savefig(f'/home/gaurav/Documents/Graphs/pid_{pid}.png')

    # Store arrays for later combined graph
    pid_arrays.append((pid, time_array, priority_array))

# Plot all 'pid's on a single graph
plt.figure()
for pid, time_array, priority_array in pid_arrays:
    plt.plot(time_array, priority_array, label=f'pid {pid}')

plt.xlabel('Time')
plt.ylabel('Priority')
plt.title('Combined Graph for all pids')
plt.legend()
# plt.savefig('/home/gaurav/Documents/Graphs/combined.png')
plt.show()

# This code creates individual graphs for each 'pid' and then combines all 'pid's into a single graph at the end. Each 'pid' has its own graph, and then all 'pid's are plotted together on a separate graph. Adjust the code according to your specific needs and preferences.


```
